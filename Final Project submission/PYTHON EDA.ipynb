```python
missing_values = df.isnull().sum()

missing_values
```

```python
df = df.dropna()

df
```

```python
df = df.drop_duplicates()

df
```

```python
no_machine_failure = df.where( df.Downtime == "No_Machine_Failure") 

no_machine_failure
```

```python
no_machine_failure.describe()
```

```python
machine_failure = df.where( df.Downtime == "Machine_Failure") 

machine_failure
```

```python
machine_failure.describe()
```

```python
comparision = machine_failure.compare(no_machine_failure)
```

```python
import matplotlib.pyplot as plt

import seaborn as sns


```

```python
df['Date'] = pd.to_datetime(df['Date'],format= 'mixed')
```

```python
df.loc[2021-12-31]
```

```python
df.hist(bins=30, figsize=(20, 15))
plt.suptitle('Histograms of Numeric Features')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(len(numeric_columns) // 3 + 1, 3, i)
    sns.boxplot(y=df[column])
    plt.title(column)
    plt.tight_layout()
plt.suptitle('Box Plots of Numeric Features', y=1.02)
plt.show()
```
